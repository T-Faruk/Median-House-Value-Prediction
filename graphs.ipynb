```python
import pandas as pd
import matplotlib.pyplot as plt

csv=pd.read_csv(r'./data/zillowHousing/clean_data_2016.csv')
csv=csv.groupby('regionidzip').mean().reset_index()
dflargest=csv.nlargest(10, ['taxvaluedollarcnt'])
templargest=pd.DataFrame()
templargest['taxvaluedollarcnt']=dflargest['taxvaluedollarcnt']
templargest['regionidzip']=dflargest['regionidzip'].astype('int64')
templargest=templargest.astype(int)
templargest.plot(x='regionidzip', y='taxvaluedollarcnt', kind='bar')
print(templargest.head())

plt.suptitle("Average Most Expensive Houses in the zipcodes")
plt.show()
```

```python
dfsmallest=csv.nsmallest(10, ['taxvaluedollarcnt'])
tempsmallest=pd.DataFrame()
tempsmallest['taxvaluedollarcnt']=dfsmallest['taxvaluedollarcnt']
tempsmallest['regionidzip']=dfsmallest['regionidzip']
tempsmallest=tempsmallest.astype(int)
tempsmallest.plot(x='regionidzip', y='taxvaluedollarcnt', kind='bar')
print(tempsmallest.head())

plt.suptitle("Average Cheapest Houses in the zipcodes")
plt.show()
```

```python
#Bar graphs for Size of Houses Vs Price of House with zipcode
dflargest=csv.nlargest(10, ['taxvaluedollarcnt'])
templargest=pd.DataFrame()
templargest['priceVsHouseSize']=dflargest['priceVsHouseSize']
templargest['regionidzip']=dflargest['regionidzip'].astype('int64')
templargest=templargest.astype(int)
templargest.plot(x='regionidzip', y='priceVsHouseSize', kind='bar')
print(templargest.head())

plt.suptitle("Average Big Houses with less prices in zipcodes")
plt.show()
```

```python
#Bar graphs for Size of Houses Vs Price of House with zipcode
dfsmallest=csv.nsmallest(10, ['taxvaluedollarcnt'])
tempsmallest=pd.DataFrame()
tempsmallest['priceVsHouseSize']=dfsmallest['priceVsHouseSize']
tempsmallest['regionidzip']=dfsmallest['regionidzip']
tempsmallest=tempsmallest.astype(int)
tempsmallest.plot(x='regionidzip', y='priceVsHouseSize', kind='bar')
print(tempsmallest.head())

plt.suptitle("Average Small Houses with Expensive prices in zipcodes")
plt.show()
```
